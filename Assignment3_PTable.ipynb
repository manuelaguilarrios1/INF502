This first function will be executed when the user inputs option 1. It will take the symbol that is given by the user, use it to find that symbol in the dictionary. It will then give the name of the symbol by using the element properties of it and give its full name, atomic number, row and column of the element. I use the the print and format fucntions to print this in a nice looking message. 

```
def option1(symbol):
    print("The full name of {} is {}. The atomic number is {}. It is located in row {} and column {}.".format(symbol,symbol_dict[symbol].name,symbol_dict[symbol].atomic_number,symbol_dict[symbol].row,symbol_dict[symbol].column))
    print("\n")
```

This next function will execute option 2. This function will take in an integer to tell it which property to show for all elements. It will then print out that specific property by using a for loop and the print and format functions. 

```
def option2(property):
    if property == 0:
        for symbol in symbol_dict:
            print("The name of {} ia {}".format(symbol,symbol_dict[symbol].name))
    elif property == 1:
        for symbol in symbol_dict:
            print("The atomic number of {} is {}".format(symbol, symbol_dict[symbol].atomic_number))
    elif property == 2:
        for symbol in symbol_dict:
            print("The row of {} is {}".format(symbol, symbol_dict[symbol].row))
    elif property == 3:
        for symbol in symbol_dict:
            print("The column of {} is {}".format(symbol, symbol_dict[symbol].column))
    print("\n")
```

This function will execute opion 3, which adds an element to the list. This function does not take any paramters and declares symbol_list and symbol_dict global variables so that anything that happens to them inside the loop, happens to them outside as well. It will prompt the user for the symbol, name, row, and column. Since their aren't really any errors that can happen since these are all the users own choices, no try and exceptions were used. It will store the the symbol in upper case to avoid case sensitivity needs when looking for a symbol. It will then append the new element as an Element type class, to the symbol list. After this is done, the symbol_dict will be recreated using the same symbol_list to symbol_dict code found in the first chunk of code. After this it will print out all the properties of that symbol so that the user can read them.

```
def option3():
    global symbol_list
    global symbol_dict
    symbol = str(input("Please enter the symbol of your new element: ")).upper()
    name = str(input("Please enter the name of "+symbol+": "))
    atomic_number = str(input("Please enter the atomic number of "+symbol+": "))
    row = str(input("Please enter the row of "+symbol+": "))
    column = str(input("Please enter the column of "+symbol+": "))
    symbol_list.append(Element(symbol, name, atomic_number, row, column))
    symbol_dict = {element.symbol: element for element in symbol_list}
    print(symbol+" has been successfully added")
    print("The full name of {} is {}. The atomic number is {}. It is located in row {} and column {}.".format(symbol,symbol_dict[symbol].name,symbol_dict[symbol].atomic_number,symbol_dict[symbol].row,symbol_dict[symbol].column))
    print("\n")
```

This function will execute option 4 which makes allows the user to edit a symbol. It uses a while loop to make sure that the symbol is in the dictionary. It will take an upper case version of the the symbol given by the user and check if it is in dictionary using the try and exception functions. If the symbol is not in the dictionary, it will give an error message and set symbol to "Error", reprompting the user to input the symbol value again. If the symbol is in the dictionary it will exit the while loop and prompt the user for the new properties it wants to change that symbol to. It will then take those properties, put them in an Element class, and update the dictionary. It will then reprint the new updated element so that the user can see. 

```
def option4():
    global symbol_dict
    symbol = "Error"
    while symbol == "Error":
        try:
            symbol = str(input("Please enter the symbol of the element you want to change: ")).upper()
            name = symbol_dict[symbol].name
        except KeyError:
            print("That symbol is not in the dictionary, please enter a valid symbol")
            symbol = "Error"
    name = str(input("Please enter the new name of "+symbol+": "))
    atomic_number = str(input("Please enter the new atomic number of " + symbol + ": "))
    row = str(input("Please enter the new row of " + symbol + ": "))
    column = str(input("Please enter the new column of " + symbol + ": "))
    symbol_dict[symbol] = Element(symbol, name, atomic_number, row, column)
    print(symbol+ "has been succesfully updated")
    print("The full name of {} is {}. The atomic number is {}. It is located in row {} and column {}.".format(symbol,symbol_dict[symbol].name,symbol_dict[symbol].atomic_number,symbol_dict[symbol].row,symbol_dict[symbol].column))
    print("\n")
```

This next chunk of code will execute any function that the user wants. It will start by initiallizing menu_choice to 0 so that we can enter the while loop. The while loop will be reiterated until menu_choice is equal to 5. if the user types a non integer it will throw an error using the try and except functions.  It will then reprompt them to enter a valid value. If the value falls outside of the valid options it will also reprompt them again. If the menu choice is equal to 1, fucntion option1 will execute. menu_coice = 1 and menu_choice = 2 will use try and catch to reprompt the user to enter valid values and menu_coice = 3 and 4 will use try and catch inside the fucntions. Once the option is completed, the user will be reprompted to enter another option. If they press 5 the code is they will leave the while loop and then finish the code. 

```
menu_choice = 0
while menu_choice != 5:
    print('Please pick one of the following options by entering the respective number:')
    print("1. See all the information that is stored about any element, by entering that element's symbol.")
    print("2. Choose a property, and see that property for each element in the table.")
    print("3. Enter a new element.")
    print("4. Change the attributes of an element, by entering the element's symbol.")
    print("5. Exit the program.")
    try:
        menu_choice = int(input("Your input: "))
    except ValueError:
        print("Whoops, you entered an invalid option!")
    while menu_choice < 1 or menu_choice > 5:
        try:
            menu_choice = input("Please enter a valid option: ")
            menu_choice = int(menu_choice)
        except (ValueError, TypeError):
            print("Whoops, you entered an invalid option!")
            menu_choice = 0
    if menu_choice == 1:
        symbol = "Error"
        while symbol == "Error":
            try:
                symbol = str(input("Please enter the symbol of the element you want to see the information about (not case sensitive): "))
                symbol = symbol.upper()
                option1(symbol)
            except KeyError:
                print("That symbol is not in the dictionary, please enter a valid symbol")
                symbol = "Error"
    elif menu_choice == 2:
        property = -1
        while property > 3 or property < 0:
            try:
                print("0. Name")
                print("1. Atomic Number")
                print("2. Row")
                print("3. Column")
                property = int(input("Please pick the number of the following property you would like to see: "))
            except ValueError:
                print("Error! Please pick a valid number between 0 to 3: ")
                property = -1
        option2(property)
    elif menu_choice == 3:
        option3()
    elif menu_choice == 4:
        option4()
```
